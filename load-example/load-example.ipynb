```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [15, 20]
```

```python
data = np.loadtxt("../_readings/Session AK at 03_09_2021, 13_01_19/2.txt", delimiter = ",")
data.shape, data.dtype
```

```python
chans = data.shape[1]
for channel in range(chans):
    plt.subplot(chans, 1, channel + 1)
    if channel == 8:
        plt.plot(data[:, channel], color = "orange")
    else:
        plt.plot(data[:, channel])
        plt.ylim(-150, 150)
```
